Import Python modules

```python
import pandas as pd
import numpy as np
from collections import defaultdict
import math
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(font_scale=1.0, style='ticks', palette='colorblind')
```

Read in mutation counts

```python
syn_counts_df = pd.read_csv('../results/syn_counts.csv')
```

Summarize the distribution of counts for all mutation types

```python
plt.figure(figsize=[8,3])
sns.boxplot(
    x='mut_type', y='counts', data=syn_counts_df.sort_values('mut_type'),
    whis=(5,95)
)
plt.yscale('log')
sns.despine()
plt.show()
```

Plot data for individual mutation types

```python
for (mut_type, data) in syn_counts_df.groupby('mut_type'):

    plt.figure(figsize=[12,3])
    sns.scatterplot(x='site', y='counts', data=data, alpha=0.5)
    plt.title(f'{mut_type}, N={len(data)}')
    sns.despine()
    plt.show()
```
