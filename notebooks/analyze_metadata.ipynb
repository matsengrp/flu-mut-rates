Import Python modules

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(font_scale=1.0, style='ticks', palette='colorblind')
```

Read in metadata

```python
df = pd.read_csv('../data/fluA.GCF_000865085.1.NC_007366.1.latest.metadata.tsv.gz', sep='\t')

df['date'] = pd.to_datetime(df['date'])
df.sort_values('date', inplace=True)

print('Number of sequences:', len(df))
print('Number missing dates', sum(df['date'].isnull()))
```

Plot the distribution of sample dates, subsetting to samples with dates after 1968.

```python
data = df[df['date'] > pd.to_datetime(1968)]
plt.figure(figsize=[4,3])
sns.histplot(x='date', data=data, bins=25)
plt.title('H3N2 HA sequences')
plt.ylabel('number of sequences')
plt.yscale('log')
sns.despine()
plt.show()
```
